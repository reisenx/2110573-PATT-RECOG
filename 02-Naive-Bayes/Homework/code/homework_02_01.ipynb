# Dependencies

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

# Constants

Define $\mu$ and $\sigma$ of both $P(x|w_1)$ and $P(x|w_2)$, also the $x$ variable range for graph visualization.

```python
# Parameters
MEAN_1, VAR_1 = 4, 2
MEAN_2, VAR_2 = 0, 2
SIGMA = np.sqrt(VAR_1)

# Range for x
X = np.linspace(-4, 8, 1000)
```

# Posterior Probability Calculation

Define prior probabilities and likelihood probabilities of both classes.

```python
# Likelihoods P(x|w)
likelihood_1 = norm.pdf(X, MEAN_1, SIGMA)
likelihood_2 = norm.pdf(X, MEAN_2, SIGMA)

# Priors
PRIOR_1 = 0.5
PRIOR_2 = 0.5
```

Calculate Evidence probability which can be calculated as following equation.
\begin{align*}
    P(x) = P(x|w_1)P(w_1) + P(x|w_2)P(w_2)
\end{align*}

```python
# Evidence P(x)
evidence = (likelihood_1 * PRIOR_1) + (likelihood_2 * PRIOR_2)
```

Calculate Posterior probability of both classes by using Bayes Theorem.
\begin{align*}
    P(w_i|x) &= \frac{P(x|w_i)P(w_i)}{P(x)}
\end{align*}

```python
# Posteriors P(w|x)
posterior1 = (likelihood_1 * PRIOR_1) / evidence
posterior2 = (likelihood_2 * PRIOR_2) / evidence
```

# Visualization

Visualize posterior probability of both classes with decision boundary line by using `matplotlib` library.

```python
# Plot
plt.figure(figsize=(10, 6))
plt.plot(X, posterior1, label=r"Happy Cat $P(w_1|x)$", color="blue", linewidth=2)
plt.plot(X, posterior2, label=r"Sad Cat $P(w_2|x)$", color="red", linewidth=2)

# Decision Boundary (Intersection)
plt.axvline(x=2, color="green", linestyle="--", label="Decision Boundary (x=2)")

# Plot Details
plt.title("Posterior Probabilities (Problem T2)", fontsize=14)
plt.xlabel("Normalized Food Amount (x)", fontsize=12)
plt.ylabel("Probability", fontsize=12)
plt.legend(fontsize=12)
plt.grid(True, alpha=0.3)
plt.tight_layout()

# Save plot
plt.savefig("figure_01.png")
```
