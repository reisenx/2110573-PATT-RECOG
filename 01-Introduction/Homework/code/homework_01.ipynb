# Dependencies

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

# Data Resources

Load data

```python
TRAIN_URL = "http://s3.amazonaws.com/assets.datacamp.com/course/Kaggle/train.csv"
TEST_URL = "http://s3.amazonaws.com/assets.datacamp.com/course/Kaggle/test.csv"

train = pd.read_csv(TRAIN_URL)
test = pd.read_csv(TEST_URL)
```

# Problem 8

What is the median age of the training set?

## Calculate Median

```python
median_age = train["Age"].median()
print(f"The median age is {median_age}.")
```

## Median Filling

Initially, there are some missing value on age column.

```python
train[train["Age"].isna()].head()
```

Fill the missing age value with the median value.

```python
train["Age"] = train["Age"].fillna(median_age)
```

## Test Dataset Validation

Make sure to fill the missing value on test dataset too.

```python
median_age = test["Age"].median()
test["Age"] = test["Age"].fillna(median_age)
```

# Problem 9

Some fields like 'Embarked' are categorical. They need to be converted to numbers first. We will represent S with 0, C with 1, and Q with 2.

## Convert Value

Convert categorical value on `Embarked` column from `S`, `C`, and `Q` to `0`, `1`, and `2` respectively.

- The first argument of `.loc[]` command is row selection
- The second argument of `.loc[]` command is column selection

```python
train.loc[train["Embarked"] == "S", "Embarked"] = 0
train.loc[train["Embarked"] == "C", "Embarked"] = 1
train.loc[train["Embarked"] == "Q", "Embarked"] = 2
```

Convert categorical value on `Sex` column from `male` and `female` to `0` and `1` respectively.

```python
train.loc[train["Sex"] == "male", "Sex"] = 0
train.loc[train["Sex"] == "female", "Sex"] = 1
```

## Calculate Mode

`mode()` returns a `Series` object rather than the single value.

```python
mode_embarked = train["Embarked"].mode()[0]
print(f"The mode of embarked is {mode_embarked}.")
```

```python
mode_sex = train["Sex"].mode()[0]
print(f"The mode of sex is {mode_sex}.")
```

## Mode Filling

Initially, there are some missing value on embarked column.

```python
train[train["Embarked"].isna()]
```

There is no missing value on sex column.

```python
train[train["Sex"].isna()]
```

Fill the missing value with mode.

```python
train["Embarked"] = train["Embarked"].fillna(mode_embarked)
train["Sex"] = train["Sex"].fillna(mode_sex)
```

## Test Dataset Preparation

Make sure to convert categorical values on test dataset.

```python
test.loc[test["Embarked"] == "S", "Embarked"] = 0
test.loc[test["Embarked"] == "C", "Embarked"] = 1
test.loc[test["Embarked"] == "Q", "Embarked"] = 2
```

```python
test.loc[test["Sex"] == "male", "Sex"] = 0
test.loc[test["Sex"] == "female", "Sex"] = 1
```

Also, make sure to fill missing value on test dataset.

```python
mode_embarked = test["Embarked"].mode()[0]
mode_sex = test["Sex"].mode()[0]

test["Embarked"] = test["Embarked"].fillna(mode_embarked)
test["Sex"] = test["Sex"].fillna(mode_sex)
```

# Problem 10

Write a linear regression classifier using gradient descent as learned in class. Use `Pclass`, `Sex`, `Age`, and `Embarked` as input features. You can extract the features from Pandas to NumPy.

## Configurations

```python
INPUT_FEATURES = ["Pclass", "Sex", "Age", "Embarked"]
OUTPUT_FEATURE = "Survived"
OUTPUT_FILENAME = "submission.csv"
ID_COLUMN = "PassengerId"

CLASSIFICATION_THRESHOLD = 0.50
```

## Linear Regression Model

Train a linear regression model from `scikit-learn` library.

```python
# Extract data from train dataset
X = train[INPUT_FEATURES]
y = train[OUTPUT_FEATURE]

# Train the linear regression model
model = LinearRegression()
model.fit(X, y)
```

Use pre-trained linear regression model from the previous step for prediction on test dataset.

```python
# Extract data from test dataset
X_test = test[INPUT_FEATURES]

# Input test dataset into pre-trained linear regression model.
raw_predictions = model.predict(X_test)
```

Create a DataFrame for the output prediction file.

```python
# Initialize empty DataFrame
submission = pd.DataFrame()

# Add passenger ID column to the DataFrame
submission[ID_COLUMN] = test[ID_COLUMN]

# Add raw prediction value to the DataFrame
submission[OUTPUT_FEATURE] = raw_predictions
```

```python
# Show the DataFrame
submission.head()
```

## Applying Threshold

- Since it is classification task, the predicted value has to be `0` or `1` only.
- If the predicted value is less than `0.5`, change them to `0`
- If the predicted value is more than or equal to `0.5`, change them to `1`

```python
submission[OUTPUT_FEATURE] = (
    submission[OUTPUT_FEATURE] >= CLASSIFICATION_THRESHOLD
).astype(int)
```

```python
submission.head()
```

## Save As CSV file

```python
submission.to_csv(OUTPUT_FILENAME, index=False)
```

# Problem 12

Try adding some higher order features to your training (${x_{1}}^{2}$, ${x_{1}}^{x_{2}}$, ...).
Does this model has better accuracy on the training set? How does it perform on the test set?

## Configurations

```python
INPUT_FEATURES = ["Pclass", "Sex", "Age", "Embarked"]
OUTPUT_FEATURE = "Survived"
OUTPUT_FILENAME = "submission_high_order.csv"
ID_COLUMN = "PassengerId"

CLASSIFICATION_THRESHOLD = 0.50
```

Use `PolynomialFeatures(degree=2)` to automatically calculate ${x_{1}}^{2}$ and ${x_{1}}{x_{2}}$ from the input features.

## Linear Regression Model

Train a linear regression model from `scikit-learn` library with high order features.

```python
# Train the linear regression model with high order features.
model = LinearRegression()
model.fit(X_high_order, y)
```

Use pre-trained linear regression model from the previous step for prediction on test dataset.

```python
# Input test dataset with high order features into pre-trained linear regression model.
raw_predictions = model.predict(X_test_high_order)
```

Create a DataFrame for the output prediction file.

```python
# Initialize empty DataFrame
submission_high_order = pd.DataFrame()

# Add passenger ID column to the DataFrame
submission_high_order[ID_COLUMN] = test[ID_COLUMN]

# Add raw prediction value to the DataFrame
submission_high_order[OUTPUT_FEATURE] = raw_predictions
```

```python
# Show the DataFrame
submission.head()
```

## Applying Threshold

- Since it is classification task, the predicted value has to be `0` or `1` only.
- If the predicted value is less than `0.5`, change them to `0`
- If the predicted value is more than or equal to `0.5`, change them to `1`

```python
submission_high_order[OUTPUT_FEATURE] = (
    submission_high_order[OUTPUT_FEATURE] >= CLASSIFICATION_THRESHOLD
).astype(int)
```
